```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('/content/data_science_job.csv')
```

```python
num = df.select_dtypes(include='number').columns
cat = df.select_dtypes(include='object').columns
```

```python
num
```

```python
cat
```

```python
num_col = SimpleImputer(strategy='mean')
cat_col = SimpleImputer(strategy='most_frequent')
```

```python
df[num] = num_col.fit_transform(df[num])
df[cat] = cat_col.fit_transform(df[cat])
```

```python
for col in num:
    Q1, Q3 = df[col].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df = df[(df[col] >= lower) & (df[col] <= upper)]
```

```python
df.describe()
```

```python
print("Missing values count:")
print(df.isnull().sum())
```

```python
for col in num:
    Q1, Q3 = df[col].quantile([0.25, 0.75])
    IQR = Q3 - Q1
    outliers = df[(df[col] < Q1 - 1.5*IQR) | (df[col] > Q3 + 1.5*IQR)]
    print(col, "outliers:", outliers.shape[0])
```
